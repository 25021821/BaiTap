```python
# W8A1
a, b = map(int, input().split())
try:
    print(f"{a / b:.2f}")
except ZeroDivisionError:
    print("Loi chia khong")
```

```python
# W8A2
n = int(input())
my_list = list(map(int, input().split()))
try:
    print(my_list[n])
except IndexError:
    print("Truy cap chi so sai")
```

```python
# W8A3
try:
    n = int(input())
    seen = set()
    while n != 1 and n not in seen:
        seen.add(n)
        n = sum(int(c) ** 2 for c in str(n))
    print("YES" if n == 1 else "NO")
except ValueError:
    print("NO")
```

```python
# W10A6
try:
    x = float(input())
    if x < 0:
        raise ValueError
    print(f"{x ** 0.5:.2f}")
except ValueError:
    print("So am")
```

```python
# W10A7
try:
    age = int(input())
    if age < 0:
        raise ValueError
    print(2025 - age)
except:
    print("Tuoi khong hop le")
```

```python
# W10A10
data = list(map(int, input().split()))
n = data[0]
arr = data[1:]
if len(arr) != n or len(set(arr)) != n:
    print("Mang khong hop le")
else:
    print(sum(arr))
```

```python
# W8A11
import re
t = int(input())
for _ in range(t):
    s = input()
    try:
        re.compile(s)
        print(True)
    except re.error:
        print(False)
```
