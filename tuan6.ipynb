```python
s = input().strip()
pairs = s.split()

result = {}

for p in pairs:
    key, value = p.split(":", 1)
    if key not in result:
        result[key] = []
    result[key].append(value)

print(result)
```

```python
nums = list(map(int, input().split()))

res = {'positives': 0, 'negatives': 0, 'zeros': 0}

for x in nums:
    if x > 0:
        res['positives'] += 1
    elif x < 0:
        res['negatives'] += 1
    else:
        res['zeros'] += 1

print(res)
```

```python
d1 = eval(input().strip())
d2 = eval(input().strip())

res = dict(d1)
for k, v in d2.items():
    res[k] = res.get(k, 0) + v

res_sorted = {k: res[k] for k in sorted(res)}
print(res_sorted)
```

```python
nums = list(map(int, input().split()))
k = int(input())

seen_pairs = set()
res = []

n = len(nums)
for i in range(n):
    for j in range(i + 1, n):
        a, b = nums[i], nums[j]
        if a + b == k:
            pair = (a, b)
            if pair not in seen_pairs:
                seen_pairs.add(pair)
                res.append(pair)

res.sort(key=lambda x: x[0])
print(res)
```

```python
nums = list(map(int, input().split()))
freq = {}
for x in nums:
    freq[x] = freq.get(x, 0) + 1

mx = max(freq.values())
cands = [x for x, c in freq.items() if c == mx]
print(min(cands))
```

```python
a = list(map(int, input().split()))
b = list(map(int, input().split()))

set_b = set(b)
res = []
seen = set()
for x in a:
    if x in set_b and x not in seen:
        seen.add(x)
        res.append(x)
print(res)
```

```python
items = input().split()
k = int(input())

d = {}
for i in range(0, len(items), 2):
    d[items[i]] = int(items[i + 1])

res = {key: val for key, val in d.items() if val > k}
print(res)
```

```python
n = int(input())
a = [list(map(int, input().split())) for _ in range(n)]

main_diag = [a[i][i] for i in range(n)]
anti_diag = [a[i][n - 1 - i] for i in range(n)]

print(main_diag)
print(anti_diag)
```
