```python
nums = list(map(int, input().split()))
result = []
seen = set()
for x in nums:
    if x not in seen:
        seen.add(x)
        result.append(x)
print(result)
```

```python
nums = list(map(int, input().split()))

result = []
running_sum = 0

for x in nums:
    running_sum += x
    result.append(running_sum)

print(result)
```

```python
t = tuple(map(int, input().split()))
k = int(input())

n = len(t)
if n == 0:
    rotated = t
else:
    k = k % n
    rotated = t[k:] + t[:k]

print(rotated)
```
