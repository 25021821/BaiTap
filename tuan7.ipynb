```python
def binary_search(arr, target):
    l, r = 0, len(arr) - 1
    while l <= r:
        mid = (l + r) // 2
        if arr[mid] == target:
            return mid
        if arr[mid] < target:
            l = mid + 1
        else:
            r = mid - 1
    return -1

arr = list(map(int, input().split()))
x = int(input())
print(binary_search(arr, x))
```

```python
arr = list(map(int, input().split()))

# Bubble sort
b = arr[:]
n = len(b)
for i in range(n):
    for j in range(n - i - 1):
        if b[j] > b[j + 1]:
            b[j], b[j + 1] = b[j + 1], b[j]

# Selection sort
s = arr[:]
n = len(s)
for i in range(n):
    mn = i
    for j in range(i + 1, n):
        if s[j] < s[mn]:
            mn = j
    s[i], s[mn] = s[mn], s[i]

print(b)   # hoặc print(s) tùy yêu cầu thầy/cô
```

```python
arr = list(map(int, input().split()))
freq = {}
best = None
best_cnt = 0

for x in arr:
    freq[x] = freq.get(x, 0) + 1
    c = freq[x]
    if c > best_cnt:
        best_cnt = c
        best = x

print(f"{best} xuat hien nhieu nhat, som nhat, {best_cnt} lan")
```

```python
from bisect import bisect_left

nums = eval(input().strip())  # ví dụ: [10, 9, 2, 5, 3, 7, 101, 18]
tails = []
for x in nums:
    i = bisect_left(tails, x)
    if i == len(tails):
        tails.append(x)
    else:
        tails[i] = x
print(len(tails))
```

```python
import heapq

intervals = eval(input().strip())  # [[l,r],...]
queries = eval(input().strip())    # [q1,q2,...]

intervals.sort(key=lambda x: x[0])
indexed = sorted([(q, i) for i, q in enumerate(queries)])

heap = []  # (length, r)
ans = [-1] * len(queries)
j = 0

for q, idx in indexed:
    while j < len(intervals) and intervals[j][0] <= q:
        l, r = intervals[j]
        heapq.heappush(heap, (r - l + 1, r))
        j += 1
    while heap and heap[0][1] < q:
        heapq.heappop(heap)
    ans[idx] = heap[0][0] if heap else -1

print(ans)
```

```python
nums = eval(input().strip())  # ví dụ: [5,2,6,1]

vals = sorted(set(nums))
rank = {v: i + 1 for i, v in enumerate(vals)}  # 1-based

BIT = [0] * (len(vals) + 2)

def add(i, v):
    while i < len(BIT):
        BIT[i] += v
        i += i & -i

def sum_(i):
    s = 0
    while i > 0:
        s += BIT[i]
        i -= i & -i
    return s

res = []
for x in reversed(nums):
    r = rank[x]
    res.append(sum_(r - 1))
    add(r, 1)

print(list(reversed(res)))
```

```python
from bisect import bisect_right, insort

nums = eval(input().strip())   # ví dụ: [3,3,9,9,5]
m = int(input().strip())

prefix = 0
best = 0
seen = [0]  # sorted list of prefix mods

for x in nums:
    prefix = (prefix + x) % m
    best = max(best, prefix)

    i = bisect_right(seen, prefix)
    if i < len(seen):
        best = max(best, (prefix - seen[i] + m) % m)

    insort(seen, prefix)

print(best)
```

```python
from collections import Counter

arr = list(map(int, input().split()))
cnt = Counter(arr)

best = 0
for x in cnt:
    if x + 1 in cnt:
        best = max(best, cnt[x] + cnt[x + 1])

print(best)
```
