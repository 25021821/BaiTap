```python
class Cylinder:
    def __init__(self, r, h):
        self.r = r
        self.h = h

    def surface_area(self):
        return 2 * 3.14 * self.r * (self.r + self.h)

    def volume(self):
        return 3.14 * self.r * self.r * self.h

r, h = map(float, input().split())
c = Cylinder(r, h)
print(f"{c.surface_area():.2f} {c.volume():.2f}")
```

```python
class Date:
    def __init__(self, d, m, y):
        self.d = d
        self.m = m
        self.y = y

    def leap(self):
        y = self.y
        return (y % 400 == 0) or (y % 4 == 0 and y % 100 != 0)

    def valid(self):
        if self.y <= 0 or not (1 <= self.m <= 12) or self.d < 1:
            return False
        mdays = [31, 29 if self.leap() else 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        return self.d <= mdays[self.m - 1]

    def next_day(self):
        mdays = [31, 29 if self.leap() else 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        d, m, y = self.d, self.m, self.y
        d += 1
        if d > mdays[m - 1]:
            d = 1
            m += 1
            if m > 12:
                m = 1
                y += 1
        return Date(d, m, y)

s = input().strip()
dd, mm, yy = map(int, s.split("/"))
dt = Date(dd, mm, yy)

if not dt.valid():
    print("INVALID")
else:
    nd = dt.next_day()
    print(f"{nd.d:02d}/{nd.m:02d}/{nd.y:04d}")
```

```python
import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def is_origin(self):
        return self.x == 0 and self.y == 0

    def on_y(self):
        return self.x == 0

    def on_x(self):
        return self.y == 0

    def dist(self):
        return math.hypot(self.x, self.y)

x, y = map(float, input().split())
p = Point(x, y)

print("ORIGIN" if p.is_origin() else "NOT ORIGIN")
print("ON Y-AXIS" if p.on_y() else "NOT ON Y-AXIS")
print("ON X-AXIS" if p.on_x() else "NOT ON X-AXIS")
print(f"{p.dist():.2f}")
```

```python
class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def add(self): return self.a + self.b
    def subtract(self): return self.a - self.b
    def product(self): return self.a * self.b
    def divide(self): return None if self.b == 0 else self.a / self.b
    def power(self): return self.a ** self.b
    def mod(self): return None if self.b == 0 else self.a % self.b
    def set_numbers(self, a, b):
        self.a = a
        self.b = b

a, b = map(float, input().split())
cal = Calculator(a, b)

while True:
    choice = input().strip()  # 1..7
    if choice == "1":
        print(cal.add())
    elif choice == "2":
        print(cal.subtract())
    elif choice == "3":
        print(cal.product())
    elif choice == "4":
        r = cal.divide()
        print("INVALID" if r is None else r)
    elif choice == "5":
        print(cal.power())
    elif choice == "6":
        r = cal.mod()
        print("INVALID" if r is None else r)
    elif choice == "7":
        a, b = map(float, input().split())
        cal.set_numbers(a, b)
        print("OK")

    out = input().strip()
    if out in ("yes", "YES", "Yes"):
        break
```

```python
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add(self, name, price):
        self.items.append({"name": name, "price": price})

    def remove(self, name):
        for i, it in enumerate(self.items):
            if it["name"] == name:
                self.items.pop(i)
                return True
        return False

    def empty(self):
        return len(self.items) == 0

    def total(self):
        return sum(it["price"] for it in self.items)

    def show(self):
        if self.empty():
            print("EMPTY")
        else:
            for it in self.items:
                print(f"{it['name']} {it['price']:.2f}")

    def clear(self):
        self.items.clear()

cart = ShoppingCart()

while True:
    cmd = input().strip()  # 2..7
    if cmd == "2":
        name = input().strip()
        price = float(input().strip())
        cart.add(name, price)
        print("OK")
    elif cmd == "3":
        name = input().strip()
        print("OK" if cart.remove(name) else "NOT FOUND")
    elif cmd == "4":
        print("YES" if cart.empty() else "NO")
    elif cmd == "5":
        print(f"{cart.total():.2f}")
    elif cmd == "6":
        cart.show()
    elif cmd == "7":
        cart.clear()
        print("OK")

    out = input().strip()
    if out in ("yes", "YES", "Yes"):
        break
```

```python
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

    def perimeter(self):
        return 2 * (self.w + self.h)

    def scale(self, k):
        self.w *= k
        self.h *= k

w, h, k = map(float, input().split())
r = Rectangle(w, h)
r.scale(k)
print(f"{r.area():.2f} {r.perimeter():.2f}")
```

```python
import math

class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError("den != 0")
        if den < 0:
            num, den = -num, -den
        g = math.gcd(num, den)
        self.num = num // g
        self.den = den // g

    def add(self, o): return Fraction(self.num * o.den + o.num * self.den, self.den * o.den)
    def sub(self, o): return Fraction(self.num * o.den - o.num * self.den, self.den * o.den)
    def mul(self, o): return Fraction(self.num * o.num, self.den * o.den)
    def div(self, o): return Fraction(self.num * o.den, self.den * o.num)

    def __str__(self):
        return f"{self.num}/{self.den}"

a, b, op, c, d = input().split()
a, b, c, d = int(a), int(b), int(c), int(d)
f1 = Fraction(a, b)
f2 = Fraction(c, d)

if op == "+":
    print(f1.add(f2))
elif op == "-":
    print(f1.sub(f2))
elif op == "*":
    print(f1.mul(f2))
else:
    print(f1.div(f2))
```

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, x):
        self.balance += x

    def withdraw(self, x):
        if x <= self.balance:
            self.balance -= x

owner, init_balance, n = input().split()
init_balance = float(init_balance)
n = int(n)

acc = BankAccount(owner, init_balance)

for _ in range(n):
    cmd, x = input().split()
    x = float(x)
    if cmd == "DEPOSIT":
        acc.deposit(x)
    else:
        acc.withdraw(x)

print(f"{acc.balance:.2f}")
```

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, subject, score):
        self.scores.append(score)

    def avg(self):
        return sum(self.scores) / len(self.scores) if self.scores else 0.0

    def rank(self):
        a = self.avg()
        if a >= 8:
            return "Excellent"
        if a >= 6.5:
            return "Good"
        if a >= 5:
            return "Average"
        return "Poor"

name = input().strip()
n = int(input().strip())
st = Student(name)

for _ in range(n):
    subject, score = input().split()
    st.add_score(subject, float(score))

a = st.avg()
print(f"{st.name} {a:.2f} {st.rank()}")
```

```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, name):
        return sum(1 for b in self.books if b.author == name)

    def find_by_year(self, y):
        return sum(1 for b in self.books if b.year == y)

lib = Library()
n = int(input().strip())

for _ in range(n):
    line = input().strip()
    if line.startswith("ADD "):
        payload = line[4:]
        title, author, year = payload.split(";")
        lib.add(Book(title, author, int(year)))
    elif line.startswith("COUNTYEAR "):
        y = int(line.split()[1])
        print(lib.find_by_year(y))
    elif line.startswith("COUNT "):
        author = line[6:]
        print(lib.count_by_author(author))
```

```python
import math

def vec(a, b):
    return (b[0]-a[0], b[1]-a[1], b[2]-a[2])

def cross(u, v):
    return (u[1]*v[2]-u[2]*v[1],
            u[2]*v[0]-u[0]*v[2],
            u[0]*v[1]-u[1]*v[0])

def dot(u, v):
    return u[0]*v[0] + u[1]*v[1] + u[2]*v[2]

def norm(u):
    return math.sqrt(dot(u, u))

A = tuple(map(float, input().split()))
B = tuple(map(float, input().split()))
C = tuple(map(float, input().split()))
D = tuple(map(float, input().split()))

n1 = cross(vec(A, B), vec(A, C))
n2 = cross(vec(B, C), vec(B, D))

cosphi = dot(n1, n2) / (norm(n1) * norm(n2))
cosphi = max(-1.0, min(1.0, cosphi))
phi = math.degrees(math.acos(cosphi))

print(f"{phi:.2f}")
```

```python
class ComplexNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def multiply(self, other):
        a, b = self.real, self.imag
        c, d = other.real, other.imag
        return ComplexNumber(a*c - b*d, a*d + b*c)

a, b = map(float, input().split())
c, d = map(float, input().split())

z1 = ComplexNumber(a, b)
z2 = ComplexNumber(c, d)
z3 = z1.multiply(z2)

print(f"{z3.real:g} {z3.imag:g}")
```

```python
import math

class Triangle:
    def __init__(self, e1, e2, e3):
        self.e1 = e1
        self.e2 = e2
        self.e3 = e3

    def getArea(self):
        a, b, c = self.e1, self.e2, self.e3
        if a <= 0 or b <= 0 or c <= 0:
            return -1
        if a + b <= c or a + c <= b or b + c <= a:
            return -1
        p = (a + b + c) / 2
        return math.sqrt(p * (p - a) * (p - b) * (p - c))

a, b, c = map(int, input().split())
t = Triangle(a, b, c)
area = t.getArea()
if area < 0:
    print("invalid")
else:
    print(f"{area:.2f}")
```
