```python
#W5A1
a, b = map(int, input().split())
def Max_Of_Two(a, b) -> int:
    return a if a > b else b
print(Max_Of_Two(a, b))
```

```python
#W5A2
def swap(a, b):
    return b, a
a, b = map(int, input().split())
x, y = swap(a, b)
print(x, y)
```

```python
#W5A3
def is_prime(n:int) -> bool:
    if n<2:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
        i+=1
    return True
n=int(input())
print(is_prime(n))
```

```python
#W5A4
def perfect_num(n: int) -> bool:
    if n <=1:
        return False
    s=0
    for i in range(1, n):
        if n%i==0:
            s+=i
            if s==n:
                return True
    return False
n=int(input())
print(perfect_num(n))
```

```python
#W5A5
def first_pos(l, k: int)-> int:
    for n, m in enumerate(l, start=1):
        if m==k:
            return n
    return -1
l=list(map(int,input().split()))
k=int(input())
print(first_pos(l,k))
```

```python
#W5A6
def fact(n:int) -> int:
    gt=1
    for i in range(1,n+1):
        gt*=i
    return gt
n=int(input())
print(fact(n))
```

```python
#W5A7
def calc(num1: float, oper: str, num2: float) -> float:
    if oper == '+':
        return num1 + num2
    if oper == '-':
        return num1 - num2
    if oper == '*':
        return num1 * num2
    return num1 / num2

a, op, b = input().split()
print(f"{calc(float(a), op, float(b)):.2f}")
```

```python
#W5A8
def hamming_distance(x: int, y: int) -> int:
    count = 0
    while x > 0 or y > 0:
        if (x & 1) != (y & 1):
            count += 1
        x >>= 1
        y >>= 1
    return count

x, y = map(int, input().split())
print(hamming_distance(x, y))
```

```python
#W5A10
a, b = input().split()

def is_isomorphic(a, b):
    map_ab = {}
    map_ba = {}

    for x, y in zip(a, b):
        if x in map_ab:
            if map_ab[x] != y:
                return False
        else:
            map_ab[x] = y

        if y in map_ba:
            if map_ba[y] != x:
                return False
        else:
            map_ba[y] = x

    return True

print(str(is_isomorphic(a, b)).lower())
```
