```python
#W4A6
def nt(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

a, b = map(int, input().split())
tong = 0
for i in range(a, b + 1):
    if nt(i):
        tong += i
print(tong)
```

```python
#W4A8
n = int(input())
count = 0

while True:
    s = str(n)
    if s == s[::-1]:
        print(count, n)
        break
    else:
        rev = int(s[::-1])
        n = n + rev
        count = count + 1
```

```python
#W4A9
n = int(input())
found = 0
i = 1
while i * i <= n:
    s = str(i * i)
    j = 0
    ok = True
    while j < len(s):
        k = j + 1
        while k < len(s):
            if s[j] == s[k]:
                ok = False
            k = k + 1
        j = j + 1
    if ok == True:
        print(i * i, end=" ")
        found = 1
    i = i + 1

if found == 0:
    print("no number")
```

```python
#W4A10
n = int(input())
maxlen = 0
num = 1
i = 1
while i <= n:
    k = i
    length = 1
    while k != 1:
        if k % 2 == 0:
            k = k // 2
        else:
            k = 3 * k + 1
        length = length + 1
    if length > maxlen:
        maxlen = length
        num = i
    i = i + 1
print(num, maxlen)
```

```python
#W4A13
a, b = map(int, input().split())
sum_a = 0
sum_b = 0
i = 1
while i < a:
    if a % i == 0:
        sum_a = sum_a + i
    i = i + 1
j = 1
while j < b:
    if b % j == 0:
        sum_b = sum_b + j
    j = j + 1
if sum_a == b and sum_b == a:
    print("true")
else:
    print("false")
```

```python
#W4A15
head, leg = map(int, input().split())

if leg % 2 != 0:
    print("Sai")
else:
    y = (leg - 2*head) / 2
    x = head - y
    if x >= 0 and y >= 0:
        print(x, y)
    else:
        print("Sai")
```

```python
#W4A27
n = input()
count = 0
in_word = False   # flag to check if we are currently inside a word
for i in range(len(n)):
    if n[i] != " ":          # letter
        if not in_word:      # start of a new word
            count += 1
            in_word = True
    else:                    # s[i] is a space
        in_word = False      # we left the word
print(count)
```

```python
#W4A30
s = input("Nhập vào một chuỗi: ")
hoa = thuong = so = 0
for ch in s:
    if ch.isupper():
        hoa += 1
    elif ch.islower():
        thuong += 1
    elif ch.isdigit():
        so += 1
print("Ký tự in hoa:", hoa)
print("Ký tự in thường:", thuong)
print("Ký tự số:", so)
```
