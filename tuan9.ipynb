```python
a = list(map(int, input().split()))
b = list(map(int, input().split()))
i = j = 0
res = []
while i < len(a) and j < len(b):
    if a[i] <= b[j]:
        res.append(a[i])
        i += 1
    else:
        res.append(b[j])
        j += 1
res.extend(a[i:])
res.extend(b[j:])
print(*res)
```

```python

# W9A2
n, m = map(int, input().split())
matrix = [list(map(int, input().split())) for _ in range(n)]
k = int(input())
arr = []
for col in range(m):
    for row in range(n):
        arr.append(matrix[row][col])
arr.sort()
print(arr[k - 1])
```

```python
# W9A3
s = list(input().strip())
best = s[:]
n = len(s)
for i in range(n):
    for j in range(i + 1, n):
        s[i], s[j] = s[j], s[i]
        if s[0] != '0' and s < best:
            best = s[:]
        s[i], s[j] = s[j], s[i]
print("".join(best))
```

```python
# W9A4
q = int(input())
for _ in range(q):
    a = input().strip()
    b = input().strip()
    i = j = 0
    while i < len(a) and j < len(b):
        if a[i].upper() == b[j]:
            j += 1
        i += 1
    print("YES" if j == len(b) else "NO")
```

```python
# W9A5
list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))
found = False
for i in range(len(list2) - len(list1) + 1):
    if list2[i:i + len(list1)] == list1:
        found = True
        break
print("YES" if found else "NO")
```

```python
# W9A6
n = int(input())
a = list(map(int, input().split()))
i = n - 2
while i >= 0 and a[i] >= a[i + 1]:
    i -= 1
if i < 0:
    a.reverse()
else:
    j = n - 1
    while a[j] <= a[i]:
        j -= 1
    a[i], a[j] = a[j], a[i]
    a[i + 1:] = reversed(a[i + 1:])
print(*a)
```

```python
# W9A7
bed = list(map(int, input().split()))
k = int(input())
count = 0
for i in range(len(bed)):
    if bed[i] == 0:
        left = (i == 0 or bed[i - 1] == 0)
        right = (i == len(bed) - 1 or bed[i + 1] == 0)
        if left and right:
            bed[i] = 1
            count += 1
print(count >= k)
```

```python
# W9A8
moves = input().strip()
x = y = 0
seen = {(0, 0)}
cross = False
for m in moves:
    if m == 'U': y += 1
    elif m == 'D': y -= 1
    elif m == 'L': x -= 1
    elif m == 'R': x += 1
    if (x, y) in seen:
        cross = True
        break
    seen.add((x, y))
print(cross)
```

```python
# W9A9
a, b, c, d, e = map(int, input().split())
if a > b: a, b = b, a
if c > d: c, d = d, c
if a > c: a, c = c, a
if b > d: b, d = d, b
if b > c: b, c = c, b
if e < b: print(b)
elif e > c: print(c)
else: print(e)
```

```python
# W9A10
houses = list(map(int, input().split()))
heaters = list(map(int, input().split()))
houses.sort()
heaters.sort()
res = 0
j = 0
for h in houses:
    while j + 1 < len(heaters) and abs(heaters[j + 1] - h) <= abs(heaters[j] - h):
        j += 1
    res = max(res, abs(heaters[j] - h))
print(res)
```
